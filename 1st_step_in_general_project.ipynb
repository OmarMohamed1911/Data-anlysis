```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv(r'D:\Data Analyst projects\General projects\ff_race_50.csv')
df.head()
```

```python
df.dropna(axis=1, inplace = True)
df.head()
```

```python
df.insert(3,'fullname',df.First + " " + df.Last)
df.head()
```

```python
df['Time'] = pd.to_timedelta(df['Time'])
df.info()
```

```python
df['TotalMinute'] = df['Time'].dt.total_seconds() / 60
df.head()
```

```python
df['TotalMinute'] = df['TotalMinute'].round().astype(int)
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns= {'Division': 'Gender', 'fullname':'FullName'}, inplace =True)
df.head()
```

```python

def get_lat_long(city, state):
    address = f"{city}, {state}"
    try:
        geolocator = Nominatim(user_agent="running", timeout=10)
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None

```

```python
df['latitude'], df['longitude'] = zip(*df.apply(lambda x: get_lat_long(x['City'],x['State']), axis=1))
```

```python
df.head()
```

```python
df['latlong'] = df['latitude'].astype(str) + ', ' + df['longitude'].astype(str)
df.head()
```
